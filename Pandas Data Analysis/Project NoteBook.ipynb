# **Data Cleaning**

```python
#Duplicated Rows Count
df.duplicated().sum()
```

```python
df.head(5)
```

```python
df_2 = df.drop(['Education_Level','Compensation_and_Benefits','VAR1'	, "VAR2" ,	"VAR3"	 , "VAR4",	"VAR5",	"VAR6"	,"VAR7",'Travel_Rate',"Work_Life_balance" ,"Post_Level"	], axis=1)
```

```python
df_2
```

```python
# Select the columns 'Employee_ID', 'Age', and 'Pay_Scale'
df_2[['Employee_ID', 'Age', 'Pay_Scale']]
```

```python
#sort data ascending by time of service

df.sorted = df_2.sort_values(by='Time_of_service', ascending=False)
df.sorted
```

```python
# prompt: create Age_bins column use Age column

df.sorted['Age_bins'] = pd.cut(df_2['Age'], bins=[18, 25, 35, 45, 55, 65], labels=['18-25', '25-35', '35-45', '45-55', '55-65'])

df.sorted
```

```python
df.sorted.groupby('Gender').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))

# Add labels to values on chart
for i, value in enumerate(df.sorted.groupby('Gender').size()):
    plt.text(value, i, str(value), ha='center', va='center', color='white', fontweight='bold', backgroundcolor='gray')

# Make gray background for labels value only
plt.gca().spines[['top', 'right']].set_visible(False)
plt.gca().yaxis.tick_left()
plt.gca().tick_params(left=False)

plt.show()
```

```python
from matplotlib import pyplot as plt
df.sorted['Time_of_service'].plot(kind='hist', bins=20, title='Time_of_service', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
df.sorted.groupby('Hometown').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# prompt: pie chart for Age pins count

age_bins_count = df.sorted['Age_bins'].value_counts()

plt.figure(figsize=(10, 7))
plt.pie(age_bins_count, labels=age_bins_count.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set2'))

plt.title('Age_bins Freq')
plt.axis('equal')
plt.show()
```

```python
# Age_bins Attrition_rate

Age_bins_att_r = df.sorted.groupby('Age_bins')['Attrition_rate'].mean()
Age_bins_att_r
```

```python
Age_bins_att_r.sort_values(ascending=False).plot(kind='bar', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
plt.ylabel = 'attrition rate'
plt.show()
```
